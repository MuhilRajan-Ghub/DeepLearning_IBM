```python
import pandas as pd
import numpy as np

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
filepath='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
concrete_data = pd.read_csv(filepath)

concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns
```

```python
predictors = concrete_data.iloc[:, :-1]
target = concrete_data.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

predictors_std = pd.DataFrame(scaler.fit_transform(predictors))
predictors_std.columns = predictors.columns
predictors_std.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
```

```python
def regression_model():
    model = Sequential()
    model.add(Input(shape=(8,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model_1 = regression_model()
model_1.fit(predictors_std, target, validation_split=0.3, epochs=100, verbose=2)
```

```python
def regression_model_2():
    model = Sequential()
    model.add(Input(shape=(8,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model_2 = regression_model_2()
model_2.fit(predictors_std, target, validation_split=0.1, epochs=100, verbose=2)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred_1 = model_1.predict(predictors_std)
y_pred_2 = model_2.predict(predictors_std)
```

```python
print('Model 1 RMSE: ', np.sqrt(mean_squared_error(target, y_pred_1)))
print('Model 1 R^2: ', r2_score(target, y_pred_1))
print('\n')
print('Model 2 RMSE: ', np.sqrt(mean_squared_error(target, y_pred_2)))
print('Model 2 R^2: ', r2_score(target, y_pred_2))
```
